```python
air = pd.read_csv('D:\Data_Science\Project\data\\raw\AirPassengers.csv')
air.head(5)
```

```python
air.info()
```

```python
air.describe()
```

```python
plt.figure(figsize = (8,3))
sns.lineplot(data = air, x = 'Month', y = 'Passengers')
plt.title('Pasangers Per Month')
plt.xlabel('Air')
plt.ylabel('Passanger')
plt.show()
```

```python
# Since the data in the x labels are being plot directly as 1949-01
# it's better to assume a month and change the oriendation of x

ax = sns.lineplot(data = air, x = 'Month', y = 'Passengers')
x_tick_labels = ax.get_xticklabels()
plt.close()
```

```python
x_tick_labels
```

```python
x_tick_labels_array = np.array(x_tick_labels)
x_tick_labels_array[range(23,144,24)]
```

```python
plt.style.use('default')
plt.figure(figsize = (8,3))
sns.lineplot(data = air, x = 'Month', y = 'Passengers')
plt.title('Pasangers Per Month')
plt.xticks(range(23,144,24))
plt.xlabel('Air')
plt.ylabel('Passanger')
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize = (10,5))
sns.lineplot(data = air, x = 'Month', y = 'Passengers', markersize = 5, 
             marker = '^', markerfacecolor = 'red')
plt.title('Pasangers Per Month')
plt.xticks(range(23,144,24))
plt.xlabel('Air')
plt.ylabel('Passanger')
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize = (10,5))
sns.lineplot(
    data = air, 
    x = 'Month', 
    y = 'Passengers', 
    markersize = 5, 
    marker = '^',
    markerfacecolor = 'red',
    lw = 1,
    ls = ':')
plt.title('Pasangers Per Month')
plt.xticks(range(23,144,24))
plt.xlabel('Air')
plt.ylabel('Passanger')
plt.ylim((0,700))
plt.show()
```
